# FITS files

```python
import matplotlib.pyplot as plt
import os
from astropy.io import fits
import numpy as np
from matplotlib import cm
```

## NGC 2903

### HI map (natural)

```python
fig=plt.figure(figsize=[12,12])
file='./data/NGC_2903_NA_MOM0_THINGS.FITS'
HI_NGC_2903_na=fits.getdata(file)

plt.imshow(np.squeeze(HI_NGC_2903_na))
plt.title('HI map data for NGS 2903', fontsize=19)
plt.savefig('./images/HI_NGC_2903_na.png',format='png')
plt.show()
```

```python
from astropy.io import fits
fig=plt.figure(figsize=[12,12])
file='./data/NGC_2903_NA_MOM0_THINGS.FITS'
HI_NGC_2903_na=fits.getdata(file)
plt.imshow(np.squeeze(HI_NGC_2903_na),origin='lower',  cmap='gist_yarg')
plt.colorbar()
plt.savefig('./images/hist_HI_NGC_2903_na.png',format='png')
plt.show()
```

### Mean intensity-weighted velocity 

```python

fig=plt.figure(figsize=[12,12])
file='./data/NGC_2903_NA_MOM1_THINGS.FITS'
velocity_HI_NGC_2903_na=fits.getdata(file)

plt.imshow(np.squeeze(velocity_HI_NGC_2903_na),origin='lower')
plt.title('HI velocity data for NGC 2903', fontsize=19)
plt.colorbar()
plt.savefig('./images/velocity_HI_NGC_2903_na.png',format='png')

plt.show()
```

### velocity dispersion

```python
file='./data/NGC_2903_NA_MOM2_THINGS.FITS'
data=fits.getdata(file)
data=np.nan_to_num(data)
data.shape
```

```python
fig=plt.figure(figsize=[12,12])
plt.imshow(np.squeeze(data),origin='lower')
plt.title('HI dispersion data for NGC 2903', fontsize=19)
plt.savefig('./images/dispersion_HI_NGC_2903_na.png',format='png')
plt.show()
```

```python
X,Y=np.meshgrid(np.arange(0,1024,1),np.arange(0,1024,1))
data = np.array(data).reshape((1024, 1024))
Z=data
Z/=1000
fig=plt.figure(figsize=[14.7,12])
picture=plt.pcolormesh(X,Y,Z)
plt.colorbar()
plt.savefig('./images/hist_dispersion_HI_NGC_2903_na.png',format='png')
plt.show()
```
